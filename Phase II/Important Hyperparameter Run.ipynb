# Analysis I

In this notebook, we have tried using analysis I, which is trying to apply GridSearchCV using every parameter and a range over every parameter to check it's effect on the evaluation metric. The thesis for the analysis is as below assumptions:
###### Assumptions
* We assume we have a set of hyperparamters and its range of values that will be turned up by the Database team.
* We then use this range to train our model and perform Cross Validation over this range by varying one hyperparameter, and fixing the rest.
* By doing so, we will be checking how is the evaluation metric changing with the increase or decrease in the hyperparameter range that is being varied.
* Now, how we are analyzing weather a hyperparameter is important or not is, if it makes a significant change on the evaluation metric, may be increase or decrease in the evaluation(drastically) - that means it's important.
3 ways to check this:

    * If the evaluation metric changes(increase or decreases drastically) - it means it is important
    * If the evaluation metric remains the same, which means no difference in it's value - it means it is not important
    * If the evaluation metric changes in a zig zag fashion, with no significant pattern - it means it is not important

Below notebook is doing somewhat similar to perform this analysis

## Analysis - Test 1
In the Test 1 - We check if by just one hyperparameter range being specified, do we get any significant change in the evaluation metric.

As our problem is a `Classification` we chosse `log loss` as our evaluation metric to be verified.

```python
model = XGBClassifier()
n_estimators = range(50, 1000, 50)
param_grid = dict(n_estimators=n_estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
grid_search = GridSearchCV(model, param_grid, scoring="neg_log_loss", n_jobs=-1, cv=kfold)
grid_result = grid_search.fit(X_train1, y_train1)
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
# plot
plt.errorbar(n_estimators, means, yerr=stds)
plt.title("XGBoost n_estimators vs Log Loss")
plt.xlabel('n_estimators')
plt.ylabel('Log Loss')
# plt.savefig('n_estimators.png')
```

#### Conclusion

Conclusion from Test - 1 is that yes, n_estimator does make difference on our evaluation metric. So, based on the graph we get at the top, the `n_estimator` is good for our model at `350`

But, that's not what we are testing. From our hypothesis, we get to know, that if the n_estimator increases after 350 it stars penalizing our model, which means, it doesn't help in any way. So, our model wont be good after that range.

## Analysis - Test 2

Now, the Test 1 having given us good result, lets go with another hyperparameter - `max_depth` being involved into the picture.
We have already concluded that `n_estimator`/`n_tree` in H2O is very important hyperparameter, we now fix `n_estimator` and vary `max_depth` to check if `max_depth` turns to be important or not.

```python
model = XGBClassifier()
max_depth = range(1, 10, 1)
n_estimators = range(349, 350, 1)
print(max_depth)
param_grid = dict(max_depth=max_depth, n_estimators=n_estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
grid_search = GridSearchCV(model, param_grid, scoring="neg_log_loss", n_jobs=-1, cv=kfold, verbose=1)
grid_result = grid_search.fit(X_train1, y_train1)
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))

plt.errorbar(max_depth, means, yerr=stds)
plt.title("XGBoost max_depth vs Log Loss")
plt.xlabel('max_depth')
plt.ylabel('Log Loss')
# plt.savefig('max_depth.png')
```

#### Conclusion

Conclusion of Test 2 - Now, as we have got the `errorplot` using Matplotlib, we can see, how `max_depth` is making difference  on the `log_loss`. As we check our hypothesis, we are checking max_depth is making any significant changes to our model evaluation metric, the answer is `Yes`. 

Thus, we conclude, even `max_depth` is important. For now, we have two major hyperparameters in our list:
    
    * n_estimator
    * max_depth
